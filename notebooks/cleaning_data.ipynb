```python
#Both Unnamed and Classification represent the same information so we are going to drop the Unnamed column
df = df.drop(columns=["Unnamed: 0"])
df = df.rename(columns={"classification": "label"})
```

```python
#The acceptt columns has 397 missing values , content_type has 43,088 missing values, length hs 43,088 missing values and content has 43, 088 missing values 
```

```python
#fixing the "Lenght: Column Name (Spelling Error)
df = df.rename(columns={"lenght": "content_length"})
```

```python
df.head()
```

```python
#HAndling Missing Values
#fix the Accept columns
df["Accept"] = df["Accept"].fillna("unknown")
```

```python
#fix the content_type 
df["content-type"] = df["content-type"].fillna("none")
```

```python
#fix the content_length
df["content_length"] = df["content_length"].fillna("0")
```

```python
df.head(100)
```

```python
#converting to Numerica values
df["content_length"] = pd.to_numeric(df["content_length"], errors="coerce")
df["content_length"] = df["content_length"].fillna(0)
```

```python
df.info()
```

```python
df["content"] = df["content"].fillna("")
```

```python
#Normalizing all text columns 
text_columns = df.select_dtypes(include="object").columns

for col in text_columns:
    df[col] = df[col].str.lower().str.strip()
```

```python
#Checking forf Duplicated rows
df.duplicated().sum()
```

```python
#Clean datasset, no missing values, no duplicates , all text is normalized , changed classification to label columns
```

```python
#Saving the Clean version of the dataset
df.to_csv("csic_cleaned.csv", index=False)
```
